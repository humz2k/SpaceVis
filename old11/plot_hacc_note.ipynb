```python
import SpaceVis
import pygio
import numpy as np 
import matplotlib.pyplot as plt
import tqdm
```

```python
plt.rcParams['figure.figsize'] = [15, 15]
plt.rcParams['figure.dpi'] = 100
```

```python
data = pygio.read_genericio("data/hacc_128.dat", ["x", "y", "z"])

coords = np.column_stack((data["x"],data["y"],data["z"]))
#coords = coords[::50]
coords = coords/128
coords *= 32
coords -= 16
```

```python
Ng = 32.0

im_x = 1080
im_y = 1080

cam_x = Ng/2
cam_y = Ng/2

cam_pos = np.array([0,0,-10],dtype=np.float32)
cam_rot = np.array([0,0,0],dtype=np.float32)
e = np.array([0,0,10],dtype=np.float32)
```

```python
out = SpaceVis.renderer.draw_frame(cam_pos,cam_rot,e,im_x,im_y,coords,50,Ng,2,2,2)
```

```python
print(np.max(out))
```

```python
plt.imshow(out,cmap="inferno",vmax=15,vmin=0)#,vmax=4,vmin=0)
plt.show()
```
